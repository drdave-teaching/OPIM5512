# 1) Clone your repo & make a branch

```python
# in Terminal/PowerShell
cd ~/Desktop                          # or wherever you work
git clone https://github.com/<you>/<your-repo>.git
cd <your-repo>
git switch -c a01/import
```

# 2) Add the professor’s repo as a read-only remote and fetch it

```python
git remote add upstream https://github.com/<prof>/<prof-repo>.git
git fetch upstream
```

# 3) Copy just the paths you want from the professor’s default branch

```python
# Pull ONLY the A01 folder into your working tree:
git checkout upstream/main -- A01-eda

# (Optional) also grab the PR template at repo root:
git checkout upstream/main -- .github/pull_request_template.md
```

# 4) Commit and push

```python
git add -A
git commit -m "A01: import starter files from professor repo"
git push -u origin a01/import
```
